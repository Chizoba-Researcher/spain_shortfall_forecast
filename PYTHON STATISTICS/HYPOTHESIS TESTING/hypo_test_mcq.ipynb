```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import pandas as pd
```

## Q.2
There are 30 employees working for the same company, that earn an average of R30 000 per month. Salaries follow a normal distribution with a standard deviation of R2000. What is the 95% confidence interval for the mean salary of the population?


```python
x = np.arange(20000, 40000, 500)
mu = 30000
sigma = 2000/(30**0.5)

f = st.norm.pdf(x, loc=mu, scale = sigma) # We now calculate f(x), the probability density function of our normal distribution.
p = st.norm.cdf(x, loc=mu, scale = sigma) # Probability for an observation x

#Confidence Interval 95%
p1 = st.norm.ppf(0.975, loc=mu, scale = sigma) # upper 95%
p2 = st.norm.ppf(0.025, loc=mu, scale = sigma) # lower 95%
print(round(p2,2), round(p1,2))

plt.plot(x, f, 'k', label ='PDF')
plt.axvline(x = p1, color = 'b', linestyle = '--', label ='Upper bound')
plt.axvline(x = p2, color = 'g', linestyle = '--', label ='lower bound')
plt.legend()
plt.show()
```

## Q7
A university wants to estimate the overall student average, taking into account 121 random student test scores. These selected test scores have an average of 67% and a standard deviation of 12%. What is the 95% confidence interval for the overall student average?

```python
x = np.arange(0.0,1.0,0.01)
n = 121
mu_x = 0.67
sigma_x = 0.12
sigma = sigma_x/(n**0.5)

#PDF
f = st.norm.pdf(x, loc=mu_x , scale = sigma_x) # We now calculate f(x), the probability density function of our normal distribution.
p = st.norm.cdf(x, loc=mu_x, scale = sigma_x) # Probability for an observation x

#Confidence Interval 95%
p1 = st.norm.ppf(0.975, loc=mu_x, scale = sigma) # upper 95%
p2 = st.norm.ppf(0.025, loc=mu_x, scale = sigma) # lower 95%
print(round(p2*100,3), round(p1*100,3))

plt.plot(x, f, 'k', label ='PDF')
plt.axvline(x = p1, color = 'b', linestyle = '--', label ='Upper bound')
plt.axvline(x = p2, color = 'g', linestyle = '--', label ='lower bound')
plt.legend()
plt.show()
```

## Q10
A food delivery service claims that they will deliver within 45 minutes. To test this claim, you place 36 food orders, timing each delivery. The average time taken by the delivery service is 48 minutes, with a standard deviation of 5 minutes. How confident are you that the food delivery service’s claim is false?

```python
mu = 45
n = 36
mu_x = 48
sigma_x = 5
sigma = sigma_x/np.sqrt(36)

# My Hypothesis statement;
# Let  H0 => mu <= 45
#      H1 => mu > 45
def mu_h(z):
    ans = (z*sigma)-mu_x
    return ans

#For CI 99%
print(f'for CI 99%:: is {-mu_h(2.58)} > 45')

#For CI 95%
print(f'for CI 95%:: is {-mu_h(1.96)} > 45')
```
